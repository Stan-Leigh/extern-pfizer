## Step 3: Build a Simple Chatbot

```python
# ============================================
# STEP 3: Import the Tools We Need
# ============================================
#
# WHAT WE'RE DOING:
# Loading the specific parts of LlamaIndex we'll use — the Gemini
# connection and the chat message format.
#
# WHY THIS MATTERS:
# Python needs you to explicitly load each tool before using it,
# like pulling specific tools out of a toolbox.
#
# WHAT YOU'LL SEE:
# No output — imports are silent unless there's an error.
# If you see "ModuleNotFoundError", go back and re-run Step 1.
# ============================================
from llama_index.llms.gemini import Gemini
from llama_index.core.llms import ChatMessage
```

```python
# ============================================
# STEP 4: Connect to the Gemini AI Model
# ============================================
#
# WHAT WE'RE DOING:
# Creating a connection to Google's Gemini 2.0 Flash model.
# This is the AI "brain" that will generate our chatbot's responses.
#
# WHY THIS MATTERS:
# Gemini Flash is fast and free — perfect for learning and experimenting.
# In a real company, you might use a more powerful (and expensive) model.
#
# WHAT YOU'LL SEE:
# No output — the model is ready and waiting for questions.
# ============================================
llm = Gemini(
    model="models/gemini-2.5-flash",
)
```

```python
# ============================================
# STEP 5: Build the Chatbot Function
# ============================================
#
# WHAT WE'RE DOING:
# Creating the main chatbot — a loop that takes your questions,
# sends them to Gemini, and prints the AI's response.
#
# WHY THIS MATTERS:
# This is the core of any chatbot! The conversation history (messages list)
# lets the AI remember what you talked about earlier in the conversation.
#
# WHAT YOU'LL SEE:
# Nothing yet — we're just defining the function. We'll run it in the next step.
# ============================================
def simple_chatbot():
    """
    A simple interactive chatbot using Gemini with LlamaIndex.
    """
    print("Simple Gemini Chatbot")
    print("Type 'exit' to end the conversation")
    print("-" * 50)

    # Initialize chat history — this stores the conversation so far
    messages = []

    while True:
        # Get user input
        user_input = input("\nYou: ")

        # Check for exit command
        if user_input.lower() in ["exit", "quit", "bye"]:
            print("\nChatbot: Goodbye! Have a great day!")
            break

        # Handle empty input
        if not user_input.strip():
            print("(Please type a question, or type 'exit' to quit.)")
            continue

        # Add user message to history
        messages.append(ChatMessage(role="user", content=user_input))

        try:
            # Send the full conversation to Gemini and get a response
            response = llm.chat(messages)

            # Print the response
            print(f"\nChatbot: {response.message.content}")

            # Add assistant response to history so it remembers this exchange
            messages.append(ChatMessage(role="assistant", content=response.message.content))

        except Exception as e:
            print(f"\nError: {e}")
            print("Please try again or check your API key")
```

```python
# ============================================
# STEP 6: Run Your Chatbot!
# ============================================
#
# WHAT WE'RE DOING:
# Starting the chatbot so you can have a conversation with Gemini.
#
# WHY THIS MATTERS:
# This is the moment of truth — you're running your first AI chatbot!
#
# WHAT YOU'LL SEE:
# A "You: " prompt where you can type questions.
# Type any question and press Enter to get a response.
# Type "exit" to stop the chatbot.
#
# TRY THESE QUESTIONS:
# - "What is artificial intelligence?"
# - "Explain RAG in simple terms"
# - "What is a pharmaceutical SDF document?"
# ============================================
if __name__ == "__main__":
    simple_chatbot()
```
